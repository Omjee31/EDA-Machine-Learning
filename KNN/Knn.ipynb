# 🎯 K-Nearest Neighbors (KNN)

---

## 🧠 What is K-Nearest Neighbors (KNN)?

> **K-Nearest Neighbors (KNN)** is a **supervised machine learning algorithm** used for:

- 📌 Classification  
- 📌 Regression  

✨ **Core Idea:**  
> It works on **distance-based learning**, where a new data point is predicted based on the labels of its nearest neighbors.

---

## ⚡ Key Concepts

- 🔢 **K** → Number of nearest neighbors  
- 📏 **Distance** → Used to measure similarity between data points  
- 📌 Most commonly used:
  > **Euclidean Distance**

---

## ❓ Why We Use KNN?

> **We use KNN when we want to make predictions based on similarity between data points.**

✅ Simple  
✅ No training phase  
✅ Works for both classification & regression  
✅ Good for small & non-linear datasets  

---

### 💡 Intuition

> 🧠 “Similar data points have similar outcomes.”

---

## 📌 When We Use KNN?

✔ Use KNN when:

1. 🔍 Pattern Recognition (Similarity-based learning)  
2. 📊 Small Dataset  
3. 📈 Non-Linear Data  
4. ⚡ No Training Required (Direct prediction)  

---

## 🚫 When We DON'T Use KNN?

❌ Avoid KNN when:

1. 📦 Large Dataset (slow performance)  
2. 📉 High-Dimensional Data (curse of dimensionality)  
3. ⚠️ Presence of Outliers  
4. 📊 Non-Homogeneous Scale (features not normalized)  
5. ⚖️ Imbalanced Dataset (e.g., 98% vs 2%)  
6. 🚀 Not suitable for real-time inference  

---

## 🌍 Where We Use KNN?

> KNN is used in real-world applications where **similarity matters**

1. 📧 Spam Detection  
2. 🏥 Medical Diagnosis  
3. 🎯 Recommendation Systems  
4. 🖼️ Image Recognition  
5. 💳 Credit Scoring / Loan Approval  
6. 🎓 Student Performance Prediction  

---

## 🔍 How KNN Works?

### Step-by-Step:

1. 🔢 Choose the value of **K**  
2. 📏 Calculate distance from all training points  
3. 📍 Select **K nearest neighbors**  

---

### 🎯 Prediction:

- 📌 **Classification** → Majority Voting  
- 📌 **Regression** → Average of neighbors  

---

## 📐 Intuition (Distance Formula)

$$
d = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}
$$

# Algorithm For KNN
----
1. **Start**
2. **Make A Class KNN**
3. **initialize Constructor __init__(self,K) K = Number of neighbours**
   > self.k = k
   > self.X_train = None
   > self.y_train = none

---

4. **Define Fit Method  def fit(self,X_train,y_train)**
   > self.X_train = X_train
   > self.y_train = y+train


----
5. **define Pedict Method def Predict(self,X_test)**
   * initialize **y_pred** = [    ] **To Store Output**
   * start Iteratiing X_test **for i in X_test**
   *  set distances = [   ] **Store Distance From Test_pint To Training Point**
   * start iterating X_train **for i in self.X_train**
   * store Distance Between test_point And train_point in another variable **dist**
      > use calculate  Distance method To Find Distance
   * **dist = self.calculate_distance(i,j)**
      > i - Test_point
      > j - train_point
      
   * **distance.append(dist)**
   *  Close Traing Loop
   * **Now finding index of Top K_nearest_neighbours after sortiing**
       1. k_index =np.argsort(distance)[:self.k]
   * **Finding Values From y_train With The Help Of index**
       1. k_values = [self.y_train[i] for i in k_index]
   * **Now Finding Majority Count Of Values And Predict Output**
        1. most_commom = Counter(k_values).most_common(1)
   * **Finally Store Most_common Item In Y_pred**
        1. y_pred.append(most_coomon)
   * **Return In Array**
       > return np.array(y_pred)

----
6. **Define Calculate Distance method def.calculate_distance(self,point_A,point_B)**
   * return np.linalg.norm(point_A - point_B)

````python
import numpy as np
from collections import Counter

class KNN:

    # K-Nearest Neighbors (KNN) Classifier
    

    def __init__(self, k=3):
       
       # Initialize KNN with number of neighbors (k)
        
        self.k = k
        self.X_train = None
        self.y_train = None

    def fit(self, X_train, y_train):
        
        # Store training data (no actual training happens in KNN)
        
        self.X_train = X_train
        self.y_train = y_train

    def predict(self, X_test):
        
        # Predict labels for test data
       
        y_pred = []

        # Loop over each test sample
        for test_point in X_test:
            distances = []

            # Compute distance from all training points
            for train_point in self.X_train:
                dist = self.calculate_distance(test_point, train_point)
                distances.append(dist)

            # Get indices of k nearest neighbors
            k_indices = np.argsort(distances)[:self.k]

            # Get labels of k nearest neighbors
            k_labels = [self.y_train[i] for i in k_indices]
            """
            Nearest neighbors ke labels nikaalte hain
            Ye list hogi jaise: [0, 1, 1, 2, 1]
            """

            # Majority voting
            most_common = Counter(k_labels).most_common(1)
            y_pred.append(most_common[0][0])

        return np.array(y_pred)

    def calculate_distance(self, A, B):
        """
        Calculate Euclidean distance between two points
        """
        return np.linalg.norm(A - B)
````
