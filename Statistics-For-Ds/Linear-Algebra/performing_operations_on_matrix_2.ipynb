```python
# Two movie embeddings are u=[0.2,0.9,0.3] and v=[0.8,0.1,0.5].
#  Compute cosine similarity. Would you recommend movie v to someone who liked movie u?
import numpy as np
u=np.array([0.2,0.9,0.3]) 
v = np.array([0.8,0.1,0.5])
cos_theta = (u.dot(v))/(np.linalg.norm(u) * np.linalg.norm(v))
print("Cosine = ",cos_theta)

```

```python
# Compute the dot product of [1,0,0] with [0,1,0]. 
# Then with [1,0,0] itself. What do these two results tell you about the vectors?


a = np.array([1,0,0])
b = np.array([0,1,0])

print("a · b =", a.dot(b))
print("a · a =", a.dot(a))
```

👉 Linearly Independent

✔️ Each vector adds a new direction
✔️ No redundancy

🔥 Rule to Remember
If rank = number of vectors → independent
If rank < number of vectors → dependent

Linear independence Are these three vectors linearly independent: [1,0,0], [0,1,0], [0,0,1]? What about [1,2,3], [2,4,6], [3,6,9]? Use np.linalg.matrix_rank to verify.

```python
import numpy as np

# Set 1
A = np.array([[1,0,0],
              [0,1,0],
              [0,0,1]])

# Set 2
B = np.array([[1,2,3],
              [2,4,6],
              [3,6,9]])

print("Rank of A:", np.linalg.matrix_rank(A))
print("Rank of B:", np.linalg.matrix_rank(B))
```
