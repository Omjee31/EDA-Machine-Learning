```python
import numpy as np

a = np.array([2, 7])
b = np.array([7, 2])

diff = b - a                         # difference vector [-5, 5] wait no → [5, -5]
print("Difference vector:", diff)    # [5 -5]

# L1 by hand
l1_manual = np.sum(np.abs(diff))
print("L1 (manual):", l1_manual)     # |5| + |-5| = 10

# L1 using NumPy  L1 Is Manhhatan
l1_numpy = np.linalg.norm(diff, ord=1)
print("L1 (numpy): ", l1_numpy)      # 10.0


# L2 is Euclidian Distance 
# L2 by hand — square each difference, sum, take sqrt
l2_manual = np.sqrt(np.sum(diff**2))
print("L2 (manual):", round(l2_manual, 4))   # sqrt(25 + 25) = 7.0711

# L2 using NumPy (default when no ord given)
l2_numpy = np.linalg.norm(diff)
print("L2 (numpy): ", round(l2_numpy, 4))    # 7.0711

# Also available directly
l2_scipy = np.sqrt((diff**2).sum())
print("L2 (sqrt sum squares):", round(l2_scipy, 4))   # 7.0711
```
