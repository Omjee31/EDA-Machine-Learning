## Random Variable
**Definition**

    A random variable is a variable whose value depends on the outcome of a random event.
    
    It converts random outcomes into numerical values.

**Example:**

>Toss a coin
    
>Head = 1
    
>Tail = 0
    
>Here the variable X represents the outcome.
    
>So X is a random variable.

**Example in Data Science**

* Random variables represent uncertain data like:

* Daily website visitors

* Stock prices

* Number of customers per day

* Number of clicks on ads


**Types**

***Discrete Random Variable***
    
    * Countable values **Example:**number of customers

***Continuous Random Variable***

* Infinite values

**Example:** height, weight, temperature

***Where Used***

* Random variables are used in:

* Probability distributions

* Machine learning models

* Risk prediction

* Simulations

* Forecasting

## 1. Mean (Average)
**Definition**

The mean is the average value of a dataset. It represents the central value of data.

**Formula**

$$\bar{x} = \frac{\sum x_i}{n} $$ 

**Where**

* 𝑥i = each data value

* 𝑛 = total number of observations

**Example**

> Data: 2, 4, 6, 8

> Mean = (2 + 4 + 6 + 8) / 4 = 5

* > When Used

> When data has no extreme outliers

> Where Used in Data Science

> Feature scaling

> Data summarization

> Average customer spending

> Average website traffic

## 2. Median
**Definition**

The median is the middle value of a dataset after sorting it.

**Steps**

Arrange data in order

Pick middle value

**Example**

Data: 2, 4, 6, 8, 10

Median = 6

If even numbers:

Data: 2, 4, 6, 8

Median = (4 + 6) / 2 = 5

**When Used**

> Used when data has outliers.

**Example:**
> Income data (very high salaries can distort mean).

> Where Used in Data Science

> Income analysis

> House price prediction

> Robust statistics

## 3. Mode
**Definition**

    The mode is the most frequently occurring value in a dataset.

**Example**

    Data:2, 3, 3, 4, 5
    
    Mode = 3

**Types**

    Unimodal → one mode
    
    Bimodal → two modes
    
    Multimodal → multiple modes
    
**When Used**

    Where Used in Data Science

    Most frequent website click

## 4. Range
**Definition**

    Range measures the spread of data.
    
    It is the difference between maximum and minimum values.

**Formula**

    Range = Max − Min

**Example**

    Data:
    2, 5, 7, 10
    
    Range = 10 − 2 = 8

**When Used**

        Used for quick estimation of variability.
        
        Where Used in Data Science
        
        Data exploration
        
        Understanding spread of values

## 5. Variance
**Definition**

    Variance measures how far data points are from the mean.
    
    It shows how much data is spread out.

**Formula**

$$ \sigma^2 = \frac{\sum (x_i - \mu)^2}{N} $$

**Where**

    𝑥i = data values
    
    𝜇 = mean
    
    N = number of values

**Example** (concept)

    If data values are very different from mean, variance is large.
    
    If data values are close to mean, variance is small.
    
    Where Used in Data Science
    
    Feature importance
    
    Risk analysis
    
    Machine learning model stability

## 6. Standard Deviation (STD)
**Definition**

    Standard deviation is the square root of variance.
    
    It measures average distance of data from the mean.

**Formula**

   $$ \sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{N}} $$

**Example**

    Small STD → data close to mean
    Large STD → data spread out

**When Used**

    Used when measuring data variability.
    
    Where Used in Data Science
    
    Data normalization
    
    Anomaly detection
    
    Risk modeling
    
    Machine learning algorithms

## 1️⃣ What is Gaussian Distribution?
**Definition**

A Gaussian distribution is a continuous probability distribution where data is symmetrically distributed around the mean.

* Most values occur near the mean, and fewer values appear as we move away from the mean.

**It forms a bell-shaped curve.**

### 2️⃣ Key Properties

    1️⃣ Symmetrical shape
    2️⃣ Mean = Median = Mode
    3️⃣ Bell-shaped curve
    4️⃣ Controlled by two parameters:
        * Mean (μ)
        * Standard deviation (σ)

## 3️⃣ Gaussian Distribution Formula

$$
f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

**Where**

    μ = mean
    
    σ = standard deviation
    
    x = variable value

## 4️⃣ Real World Examples
