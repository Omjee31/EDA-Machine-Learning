```python
df1 = df
df1 = df1.drop(columns=["student_id"],axis = 1)
```

```python
cols = ["study_hours", "self_study_hours", "online_classes_hours"]

df1["Total_study_hours"] = df1[cols].sum(axis=1)
df1 = df1.drop(cols,axis = 1)
```

```python
df1.head()
```

```python
df1["entertainment_hours"] = df1[["gaming_hours","social_media_hours"]].sum(axis=1)
```

```python
df1 = df1.drop(columns=["gaming_hours","social_media_hours"])
```

```python
df1.head()
```

```python
df1.describe()
```

```python
categorical_data = [i for i in df1.columns if df1[i].dtype=="object"]
categorical_data
```

```python
numeric_cols = [i for i in df1.columns if df1[i].dtype!="object"]
numeric_cols
```

```python

discrete_numerical_data = [i for i in numeric_cols if len(df1[i].unique())<=10]
discrete_numerical_data
```

```python

continous_numerical_data = [i for i in numeric_cols if len(df1[i].unique())>16]
continous_numerical_data
```

## Data Visualisation

```python
# categorical data
for i in categorical_data:
    sns.countplot(x=df1[i])
    plt.xlabel(i)
    plt.ylabel("Count")
    plt.show()
```

```python
# discrete data
for col in discrete_numerical_data:
    sns.countplot(x = df1[col],palette = "viridis")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title(f"ContPlot of {col}")
    plt.show()
```

```python
# for continous data

for col in continous_numerical_data:
    sns.histplot(df1[col],kde = True)
    plt.xlabel(col)
    plt.ylabel("count")
    plt.title(f"Hist Plot Of {col}")
    plt.show()
```

```python
# cheaking Outliers
for col in continous_numerical_data:
    sns.boxplot(data = df1, y = col)
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
```

```python
le = LabelEncoder()
for i in categorical_data:
    df1[i] = le.fit_transform(df1[i])
```

```python
df1.head()
```

```python
df1.to_csv("cleaned_data.csv", index=False)
```

```python
X = df1.drop("exam_score",axis = 1)
y = df1["exam_score"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python


mse = mean_squared_error(y_test, y_pred)

rmse = mean_squared_error(y_test, y_pred, squared=False)

r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
```

```python


mse = mean_squared_error(y_test, y_pred_rf)

rmse = mean_squared_error(y_test, y_pred_rf, squared=False)

r2 = r2_score(y_test, y_pred_rf)

print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
from sklearn.metrics import mean_squared_error,root_mean_squared_error, r2_score

models = {
    "Random Forest": RandomForestRegressor(),
    "XGBoost": XGBRegressor(),
    "Linear Regression" : LinearRegression()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    print(name)
    print("RMSE:",root_mean_squared_error(y_test, pred))
    print("R2:", r2_score(y_test, pred))
    print()
```
