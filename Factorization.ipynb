### Matrix Decomposition (Matrix factirization methods)

Many Complex Matrix problems Are Not Solved efficently or with stability 
using the limited  precison of computers. So Matrix Decomposition are the method 
Which Reduce The matrix into constituent parts that make it easier to calculate

---
### What is matrix Decomposition
 This is a Way of reducing a matrix into its contituent parts , it is an approach 
 which can simplify more complex problems 

 - Ex :-  
     A common analogy of matrix decomposition is factoring of number such as 
     factoring of 10 = 2x5

---
### LU decomposition
many way to decompose a matrix , hence there are a range of diffrent matrix decomposition
technique. Two simple and widely used matrix decomposition methods are the Lu Matrix
And QR matrix decomposition

---

## Lu decomposition

This matrix is for Square matrix and it decomposes a matrix into **L and U components**

---
-   ### A = L.U
-   **Here A is Square matrix Which we Wish to decompose**
-   **L is Lower Triangular Matrix**
-   **U is Upper Triangular Matrix**

- > The LU decomposition is found using an iterative numerical process and can fail foor those
  > matrix that cannot decompose easily . Than comes LUP(Lu decomposition With Partial Pivotiing)
  > -  **This is more stable**
  > -  **A = L.U.P**

---
## QR Decomposition
The decomposition for (n x m) matrix  and decomposess a matrix into Q and r components 

- ### A =  Q . R
- **Where A is matrix We Want to decompose**
- **Q is M xm matrix**
- **R is Upper Triangular Matrix**

----
## Cholesky decomposition

This decomposition Is for Symmetric Square matrix Where All Values Are greater than zero, so called positive definite matrices.

- ### > A = L . L^T
-  **L is Lower Triangular Matrix  And L^t is Transpose Of L**
-  **This is Also Written In A = U . U^T**

```python
import numpy as np 
from scipy.linalg import lu
A = np.random.randint(1,10,size = [3,3])
print(A)
```

```python
P,L,U = lu(A)
print("Pivot : \n",P)
print("\nLower Triangular: \n",L)
print("\nUpper Triangular: \n",U)
```

```python
LPU =P.dot(L).dot(U)
print(LPU)
```

# QR Decomposition

```python
from numpy.linalg import qr

Q,R = qr(A,'complete')
print(Q)
print("\n",R)
```

```python
B = Q.dot(R)
print(B)
```

# Cholesky decomposition

```python
from numpy.linalg import cholesky
A = np.array([
     [2,1,1],
     [1,2,1],
     [1,1,2]])
print(A)
#factorize
L = cholesky(A)
print(L)
#reconstruct
B = L.dot(L.T)
print(B)
```
