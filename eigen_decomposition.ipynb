```python
# recunstrcutiing matrix
import numpy as np

A = np.random.randint(1,10,size=(3,3))
values,vectors = np.linalg.eig(A)
print(A)
Q = vectors
# create inverse matrix
R = np.linalg.inv(Q)
# makeing eigenvalues to diagonal matrix 
L = np.diag(values)
#reconstruct

B = Q.dot(L).dot(R)
```
